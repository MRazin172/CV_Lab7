```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
original_image = cv2.imread('/content/2a6df40571e3af407bf226dd4df0597d.jpg')
```

```python
def add_salt_pepper_noise(original_image, amount=0.005, s_vs_p=0.5):
    out = np.copy(original_image)

    num_salt = np.ceil(amount * original_image.size * s_vs_p)
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in original_image.shape]
    out[coords[0], coords[1], :] = 255

    num_pepper = np.ceil(amount * original_image.size * (1. - s_vs_p))
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in original_image.shape]
    out[coords[0], coords[1], :] = 0

    return out
```

here we are adding salt and pepper noise to the image. salt and pepper noise appears as small balck grains in the image.

this makes the images look a bit unclear.

```python
noisy_image = add_salt_pepper_noise(original_image, amount=0.01, s_vs_p=0.5)
```

```python
filtered_image = cv2.medianBlur(noisy_image, 3)
```

here we are filtering out the salt and pepper noise that we added before. we are doinf this using median blur where we blur the intesnties of the pixels using its surrounding pixel median value.

this eliminates the black grains and returns us the original image.

```python
plt.figure(figsize=(8, 5))
plt.subplot(121)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(noisy_image)
plt.title('Noisy Image')
plt.axis('off')
plt.show()

plt.figure(figsize=(5, 5))
plt.imshow(filtered_image)
plt.title('Filtered Image')
plt.axis('off')
plt.show()
```
